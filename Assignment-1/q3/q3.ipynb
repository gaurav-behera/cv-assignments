```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch
import cv2
import wandb
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
```

## Dataset Analysis and Preprocessing

```python
# load dataset
train_data_csv = pd.read_csv('dataset/train.csv')
test_data_csv = pd.read_csv('dataset/test.csv')
# sample random 100000 rows of train for val
val_data_csv = train_data_csv.sample(n=10000)
train_data_csv = train_data_csv.drop(val_data_csv.index)
```

## Model Implementation and Training

### Input Features and Dataloaders

```python
# flattened raw images
def get_features_1(images):
    return images.reshape(images.shape[0], -1).astype(np.float32) / 255

# canny edge detection
def get_features_2(images):
    images_uint8 = images.astype(np.uint8)
    canny_images = [cv2.Canny(img, 100, 200).astype(np.float32)/255 for img in images_uint8]
    return np.stack(canny_images).reshape(len(canny_images), -1)

# hog features
def get_features_3(images):
    hog = cv2.hog = cv2.HOGDescriptor(_winSize=(28,28), _blockSize=(14,14), _blockStride=(7,7), _cellSize=(7,7), _nbins=9)
    images_uint8 = images.astype(np.uint8)
    hog_features = [hog.compute(img).flatten().astype(np.float32) for img in images_uint8]
    return np.vstack(hog_features)
```

```python
# get features
train_features_1 = get_features_1(train_images)
val_features_1 = get_features_1(val_images)
test_features_1 = get_features_1(test_images)

train_features_2 = get_features_2(train_images)
val_features_2 = get_features_2(val_images)
test_features_2 = get_features_2(test_images)

train_features_3 = get_features_3(train_images)
val_features_3 = get_features_3(val_images)
test_features_3 = get_features_3(test_images)

print("Feature 1: ", train_features_1.shape[1], "values")
print("Feature 2: ", train_features_2.shape[1], "values")
print("Feature 3: ", train_features_3.shape[1], "values")
```

```python
# convert the labels to one hot encoding with 10 classes
def one_hot_encoding(labels):
    return np.eye(10)[labels]

train_labels_one_hot = one_hot_encoding(train_labels).astype(np.float32)
val_labels_one_hot = one_hot_encoding(val_labels).astype(np.float32)
test_labels_one_hot = one_hot_encoding(test_labels).astype(np.float32)
```

```python
# dataset class
class FashionDataset(torch.utils.data.Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels
    
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        return self.images[idx], self.labels[idx]
    
# datasets
train_dataset_1 = FashionDataset(train_features_1, train_labels_one_hot)
val_dataset_1 = FashionDataset(val_features_1, val_labels_one_hot)
test_dataset_1 = FashionDataset(test_features_1, test_labels_one_hot)

train_dataset_2 = FashionDataset(train_features_2, train_labels_one_hot)
val_dataset_2 = FashionDataset(val_features_2, val_labels_one_hot)
test_dataset_2 = FashionDataset(test_features_2, test_labels_one_hot)

train_dataset_3 = FashionDataset(train_features_3, train_labels_one_hot)
val_dataset_3 = FashionDataset(val_features_3, val_labels_one_hot)
test_dataset_3 = FashionDataset(test_features_3, test_labels_one_hot)

# dataloaders
train_dataloader_1 = torch.utils.data.DataLoader(train_dataset_1, batch_size=32, shuffle=True)
val_dataloader_1 = torch.utils.data.DataLoader(val_dataset_1, batch_size=32, shuffle=False)
test_dataloader_1 = torch.utils.data.DataLoader(test_dataset_1, batch_size=32, shuffle=False)

train_dataloader_2 = torch.utils.data.DataLoader(train_dataset_2, batch_size=32, shuffle=True)
val_dataloader_2 = torch.utils.data.DataLoader(val_dataset_2, batch_size=32, shuffle=False)
test_dataloader_2 = torch.utils.data.DataLoader(test_dataset_2, batch_size=32, shuffle=False)

train_dataloader_3 = torch.utils.data.DataLoader(train_dataset_3, batch_size=32, shuffle=True)
val_dataloader_3 = torch.utils.data.DataLoader(val_dataset_3, batch_size=32, shuffle=False)
test_dataloader_3 = torch.utils.data.DataLoader(test_dataset_3, batch_size=32, shuffle=False)
```

### Model Architecture

```python
# MLP model
class MLP(torch.nn.Module):
    def __init__(
        self,
        lr=0.001,
        dropout=0,
        input_dim=784,
        num_hidden_layers=0,
        hidden_dims=[],
        output_dim=10,
        activation=torch.nn.ReLU(),
        epochs=50,
        optimizer=torch.optim.Adam,
        log_wandb=False,
    ):
        super(MLP, self).__init__()
        self.lr = lr
        self.dropout = dropout
        self.input_dim = input_dim
        self.num_hidden_layers = num_hidden_layers
        self.hidden_dims = hidden_dims
        self.output_dim = output_dim
        self.activation = activation
        self.epochs = epochs
        self.optimizer = optimizer
        self.log_wandb = log_wandb

        # create model
        self.make_model()

        # determine device
        if torch.cuda.is_available():
            device = torch.device("cuda")
        else:
            device = torch.device("cpu")

        print("Device: ", device)
        self.device = device
        self.to(device)

        # adding the optimizer
        self.optimizer = self.optimizer(self.model.parameters(), lr=self.lr)
        self.criterion = torch.nn.CrossEntropyLoss()

    def make_model(self):
        """
        Create the model architecture
        """
        self.model = torch.nn.Sequential()
        dims = [self.input_dim] + self.hidden_dims + [self.output_dim]
        for i in range(len(dims) - 1):
            self.model.add_module(f"linear_{i}", torch.nn.Linear(dims[i], dims[i + 1]))
            if i != len(dims) - 2:
                self.model.add_module(f"activation_{i}", self.activation)
                self.model.add_module(f"dropout_{i}", torch.nn.Dropout(self.dropout))

    def forward(self, x):
        """Forward pass"""
        return self.model(x)

    def fit(self, train_loader, val_loader=None):
        """Train the model"""
        prev_loss = np.inf
        convergence_counter = 0
        prev_val_loss = np.inf
        patience = 0
        for epoch in range(self.epochs):
            self.train()
            running_loss = 0.0
            for images, labels in train_loader:
                images, labels = images.to(self.device), labels.to(self.device)
                self.optimizer.zero_grad()
                outputs = self(images)
                loss = self.criterion(outputs, labels)
                loss.backward()
                self.optimizer.step()
                running_loss += loss.item()

            avg_loss = running_loss / len(train_loader)
            print(f"Epoch {epoch+1}/{self.epochs}")
            if val_loader is not None:
                self.evaluate(train_loader, type="Train")
                v_loss, _ = self.evaluate(val_loader, type="Val")
                
            if avg_loss < prev_loss and prev_loss - avg_loss < 0.01:
                convergence_counter += 1
                if convergence_counter == 2:
                    break
            else:
                convergence_counter = 0
                
            if v_loss > prev_val_loss:
                patience += 1
                if patience == 2:
                    break
            else:
                patience = 0
            
            prev_loss = avg_loss
            prev_val_loss = v_loss

    def evaluate(self, loader, type, print_output=True):
        """Evaluate the model"""
        self.eval()
        correct = 0
        total = 0
        total_loss = 0
        with torch.no_grad():
            for images, labels in loader:
                images, labels = images.to(self.device), labels.to(self.device)
                outputs = self(images)
                loss = self.criterion(outputs, labels)
                total_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                _, true_labels = torch.max(labels.data, 1)
                total += labels.size(0)
                correct += (predicted == true_labels).sum().item()

        avg_loss = total_loss / len(loader)
        accuracy = 100 * correct / total
        if print_output:
            print(f"\t{type} Loss: {avg_loss}, {type} Accuracy: {accuracy}")

        if self.log_wandb:
            wandb.log({f"{type} Loss": avg_loss, f"{type} Accuracy": accuracy})
        return avg_loss, accuracy
```

### Sample Run - Feature 1

```python

def show_confusion_matrix(confusion_matrix, title = "Confusion Matrix"):
    plt.figure(figsize=(5, 5))
    plt.imshow(confusion_matrix, cmap=plt.get_cmap('Greens'))
    plt.colorbar()
    # show numbers in the confusion matrix
    for i in range(10):
        for j in range(10):
            plt.text(j, i, confusion_matrix[i, j], ha='center', va='center')
    plt.title(title)
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.xticks(range(10), labels_map.values(), rotation=90)
    plt.yticks(range(10), labels_map.values())
    plt.show()


def evaluate_model(model, test_features, test_labels_one_hot):
    preds = model(torch.tensor(test_features).to(model.device)).to("cpu").detach().numpy()
    _, predicted = torch.max(torch.tensor(preds), 1)
    _, true_labels = torch.max(torch.tensor(test_labels_one_hot), 1)
    
    print("Accuracy: ", accuracy_score(true_labels, predicted))
    print("Micro Precision: ", precision_score(true_labels, predicted, average='micro'))
    print("Micro Recall: ", recall_score(true_labels, predicted, average='micro'))
    print("Micro F1 score: ", f1_score(true_labels, predicted, average='micro'))
    print("Macro Precision: ", precision_score(true_labels, predicted, average='macro'))
    print("Macro Recall: ", recall_score(true_labels, predicted, average='macro'))
    print("Macro F1 score: ", f1_score(true_labels, predicted, average='macro'))
    cm = confusion_matrix(true_labels, predicted)
    show_confusion_matrix(cm)

def get_confusion_matrix(model, test_features, test_labels_one_hot):
    preds = model(torch.tensor(test_features).to(model.device)).to("cpu").detach().numpy()
    _, predicted = torch.max(torch.tensor(preds), 1)
    _, true_labels = torch.max(torch.tensor(test_labels_one_hot), 1)
    cm = confusion_matrix(true_labels, predicted)
    return cm
```

```python
model = MLP(epochs=5, num_hidden_layers=2, hidden_dims=[128, 64], log_wandb=False)
model.fit(train_dataloader_1, val_dataloader_1)
model.evaluate(test_dataloader_1, type="Test")
print("Results for sample run using feature 1")
evaluate_model(model, test_features_1, test_labels_one_hot)
    
```

### Hyperparameter Tuning

```python
sweep_config = {
    "method": "grid",
    "metric": {"name": "accuracy", "goal": "maximize"},
    "parameters": {
        "dropout": {"values": [0, 0.1]},
        "num_hidden_layers": {"values": [2, 3]},
        "activation": {
            "values": ["relu", "sigmoid", "tanh"]
        },
    },
}

activation_map = { "relu": torch.nn.ReLU(), "sigmoid": torch.nn.Sigmoid(), "tanh": torch.nn.Tanh() }
```

#### Feature 1

```python
# Hyperparameter tuning for feature 1
run_id = 0
models_1 = []
def train():
    global run_id
    wandb.init(project="mlp_features_1", name=f"run_{run_id}")
    run_id += 1
    config = wandb.config
    model = MLP(
        dropout=config.dropout,
        num_hidden_layers=config.num_hidden_layers,
        hidden_dims=[512, 256, 64][: config.num_hidden_layers],
        activation=activation_map[config.activation],
        log_wandb=True,
    )
    model.fit(train_dataloader_1, val_dataloader_1)
    loss, accuracy = model.evaluate(test_dataloader_1, type="Test")
    models_1.append(model)
    wandb.log({"loss": loss, "accuracy": accuracy})
    wandb.finish()


sweep_id = wandb.sweep(sweep_config, project="mlp_features_1")
wandb.agent(sweep_id, function=train)
```

**Model 1 (raw images):** Achieved the highest accuracy at 0.8954. Raw pixel data from the Fashion MNIST images likely contains rich visual information that allows the model to effectively distinguish between different clothing categories.

**Model 2 (Canny edge images):** Performed the worst, with an accuracy of 0.8557. Reducing the images to just edges likely resulted in a significant loss of discriminative information compared to the full pixel data.

**Model 3 (HOG features):** Achieved an accuracy of 0.8827, in between the other two models. HOG features capture gradient information, which seems to be more informative for clothing classification than just edges, but less so than the full pixel data.

In general, the raw image data (Model 1) appears to be the most helpful input feature for this clothing classification task. The pixel-level details and overall visual characteristics of the clothing items seem to provide the model with the richest and most comprehensive information for accurately distinguishing between the different clothing categories.

Features that focus on specific low-level visual properties, like edges (Model 2) or gradients (Model 3), while still informative, tend to lose some of the holistic visual context that is important for this type of classification problem. They can still achieve reasonably good performance, but don't match the accuracy of the full pixel-level data.

### Features impact on MLP

Raw Images (Model 1):
- The full pixel data provides the MLP with the richest visual information, allowing it to learn complex patterns and generalize well.
- However, raw images also contain a lot of redundant or irrelevant information that can slow down training and make the model more prone to overfitting.

Canny Edges (Model 2): 
- Edge features emphasize important contours and boundaries, but discard much of the contextual information.
- This can make the task more challenging for the MLP, resulting in lower overall performance compared to raw images.
- However, edge features may be more robust to variations in lighting, background, etc, and could generalize better in certain scenarios.

HOG Features (Model 3):
- Gradient-based features like HOG capture important shape information while being more compact and robust than raw pixels.
- This can help the MLP train more efficiently and avoid overfitting, potentially leading to better generalization.
- But HOG also loses some of the fine-grained visual details present in the original images.

In general, specialized feature extractors like edges (Model 2) or gradients (Model 3) can improve efficiency and robustness, but may sacrifice some accuracy compared to raw pixels.
